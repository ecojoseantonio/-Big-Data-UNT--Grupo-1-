```python
# Importar bibliotecas necesarias
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

# Suponiendo que ya tienes un DataFrame llamado df con la variable objetivo 'DESOCUPADOS'
# Dividimos en variables independientes (X) y dependiente (y)
X = df.drop('DESOCUPADOS', axis=1)  # Todas las columnas menos 'DESOCUPADOS'
y = df['DESOCUPADOS']

# Separar en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear un Pipeline para manejar el preprocesamiento y el modelo
# Imputar valores faltantes y luego entrenar el modelo con GridSearchCV
pipeline_lasso = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),  # Imputar valores NaN con la media de cada columna
    ('model', LogisticRegression(penalty='l1', solver='liblinear'))
])

pipeline_ridge = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),  # Imputar valores NaN con la media de cada columna
    ('model', LogisticRegression(penalty='l2', solver='liblinear'))
])

# Definir los hiperparámetros para cada modelo
param_grid_lasso = {'model__C': np.logspace(-4, 4, 10)}
param_grid_ridge = {'model__C': np.logspace(-4, 4, 10)}

# Modelo con penalización Lasso
lasso_model = GridSearchCV(pipeline_lasso, param_grid_lasso, cv=5, scoring='accuracy')
lasso_model.fit(X_train, y_train)

# Modelo con penalización Ridge
ridge_model = GridSearchCV(pipeline_ridge, param_grid_ridge, cv=5, scoring='accuracy')
ridge_model.fit(X_train, y_train)

# Predicciones y evaluación
y_pred_lasso = lasso_model.predict(X_test)
y_pred_ridge = ridge_model.predict(X_test)

# Resultados
print("Modelo Lasso:")
print("Mejores hiperparámetros:", lasso_model.best_params_)
print("Exactitud en test:", accuracy_score(y_test, y_pred_lasso))
print("Reporte de clasificación:\n", classification_report(y_test, y_pred_lasso))
print("Matriz de confusión:\n", confusion_matrix(y_test, y_pred_lasso))

print("\nModelo Ridge:")
print("Mejores hiperparámetros:", ridge_model.best_params_)
print("Exactitud en test:", accuracy_score(y_test, y_pred_ridge))
print("Reporte de clasificación:\n", classification_report(y_test, y_pred_ridge))
print("Matriz de confusión:\n", confusion_matrix(y_test, y_pred_ridge))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Obtener las probabilidades de clase para el modelo Lasso y Ridge
y_pred_proba_lasso = lasso_model.decision_function(X_test)
y_pred_proba_ridge = ridge_model.decision_function(X_test)

# Calcular la curva ROC y el AUC para el modelo Lasso
fpr_lasso, tpr_lasso, _ = roc_curve(y_test, y_pred_proba_lasso)
auc_lasso = auc(fpr_lasso, tpr_lasso)

# Calcular la curva ROC y el AUC para el modelo Ridge
fpr_ridge, tpr_ridge, _ = roc_curve(y_test, y_pred_proba_ridge)
auc_ridge = auc(fpr_ridge, tpr_ridge)

# Graficar las curvas ROC
plt.figure(figsize=(10, 6))
plt.plot(fpr_lasso, tpr_lasso, color='blue', label=f'Lasso (AUC = {auc_lasso:.2f})')
plt.plot(fpr_ridge, tpr_ridge, color='green', label=f'Ridge (AUC = {auc_ridge:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')

# Personalización de la gráfica
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Curvas ROC para Modelos Lasso y Ridge')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Obtener el mejor modelo de cada GridSearchCV y acceder al modelo dentro del pipeline
best_lasso_model = lasso_model.best_estimator_.named_steps['model']
best_ridge_model = ridge_model.best_estimator_.named_steps['model']

# Obtener los coeficientes para cada modelo
coef_lasso = best_lasso_model.coef_.flatten()
coef_ridge = best_ridge_model.coef_.flatten()

# Nombres de las variables
feature_names = X_test.columns if hasattr(X_test, 'columns') else [f'Var{i}' for i in range(len(coef_lasso))]

# Ordenar los coeficientes por valor absoluto para una mejor visualización
indices_lasso = np.argsort(np.abs(coef_lasso))[::-1]
indices_ridge = np.argsort(np.abs(coef_ridge))[::-1]

# Graficar los coeficientes del modelo Lasso
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.barh(np.array(feature_names)[indices_lasso][:10], coef_lasso[indices_lasso][:10], color='blue')
plt.xlabel('Coeficiente')
plt.title('Principales Variables (Lasso)')

# Graficar los coeficientes del modelo Ridge
plt.subplot(1, 2, 2)
plt.barh(np.array(feature_names)[indices_ridge][:10], coef_ridge[indices_ridge][:10], color='green')
plt.xlabel('Coeficiente')
plt.title('Principales Variables (Ridge)')

plt.tight_layout()
plt.show()

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    confusion_matrix, roc_curve, auc, 
    accuracy_score, roc_auc_score, ConfusionMatrixDisplay
)
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer  # Para imputar valores faltantes
import matplotlib.pyplot as plt

# Asegúrate de que ya tienes X_train, X_test, y_train, y_test

# Paso 1: Imputación de valores faltantes (NaN)
imputer = SimpleImputer(strategy='mean')  # Reemplaza NaN con la media de cada columna
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)

# Paso 2: Implementar los modelos con penalización L1 y L2
model_l1 = LogisticRegression(penalty='l1', solver='liblinear', C=1.0)  # Lasso
model_l2 = LogisticRegression(penalty='l2', solver='liblinear', C=1.0)  # Ridge

# Paso 3: Ajustar los modelos
model_l1.fit(X_train, y_train)
model_l2.fit(X_train, y_train)

# Paso 4: Evaluar y mostrar métricas para ambos modelos
for model, name in zip([model_l1, model_l2], ["L1 (Lasso)", "L2 (Ridge)"]):
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]

    # Matriz de confusión
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.title(f"Matriz de Confusión - {name}")
    plt.show()

    # Curva ROC y AUC
    fpr, tpr, thresholds = roc_curve(y_test, y_proba)
    roc_auc = roc_auc_score(y_test, y_proba)

    plt.figure()
    plt.plot(fpr, tpr, label=f"Curva ROC (AUC = {roc_auc:.2f}) - {name}")
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('Tasa de Falsos Positivos (FPR)')
    plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
    plt.title(f"Curva ROC - {name}")
    plt.legend(loc="lower right")
    plt.show()

    # Exactitud
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{name} - Exactitud: {accuracy:.4f}")
    print(f"{name} - AUC: {roc_auc:.4f}")

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    confusion_matrix, roc_curve, auc, 
    accuracy_score, roc_auc_score, ConfusionMatrixDisplay
)
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt


# Filtrar los datos por año
df_2004 = df[df['ANO4'] == 2004]
df_2024 = df[df['ANO4'] == 2024]

# Separar variables predictoras (X) y la variable objetivo (y)
X_2004, y_2004 = df_2004.drop(columns=['ANO4', 'DESOCUPADOS']), df_2004['DESOCUPADOS']
X_2024, y_2024 = df_2024.drop(columns=['ANO4', 'DESOCUPADOS']), df_2024['DESOCUPADOS']



# Imputación de valores faltantes
imputer = SimpleImputer(strategy='mean')

# Aplicar imputación a los conjuntos de 2004 y 2024
X_2004 = imputer.fit_transform(X_2004)
X_2024 = imputer.fit_transform(X_2024)

# Dividir los datos en entrenamiento y prueba
X_train_2004, X_test_2004, y_train_2004, y_test_2004 = train_test_split(X_2004, y_2004, test_size=0.3, random_state=42)
X_train_2024, X_test_2024, y_train_2024, y_test_2024 = train_test_split(X_2024, y_2024, test_size=0.3, random_state=42)

# Crear modelos para ambos años
def entrenar_y_evaluar_modelo(X_train, X_test, y_train, y_test, año):
    # Modelos con penalización L1 y L2
    model_l1 = LogisticRegression(penalty='l1', solver='liblinear', C=1.0)
    model_l2 = LogisticRegression(penalty='l2', solver='liblinear', C=1.0)

    # Entrenar los modelos
    model_l1.fit(X_train, y_train)
    model_l2.fit(X_train, y_train)

    # Evaluar los modelos
    for model, name in zip([model_l1, model_l2], ["L1 (Lasso)", "L2 (Ridge)"]):
        y_pred = model.predict(X_test)
        y_proba = model.predict_proba(X_test)[:, 1]

        # Matriz de confusión
        cm = confusion_matrix(y_test, y_pred)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm)
        disp.plot()
        plt.title(f"Matriz de Confusión - {name} - Año {año}")
        plt.show()

        # Curva ROC y AUC
        fpr, tpr, thresholds = roc_curve(y_test, y_proba)
        roc_auc = roc_auc_score(y_test, y_proba)

        plt.figure()
        plt.plot(fpr, tpr, label=f"Curva ROC (AUC = {roc_auc:.2f}) - {name}")
        plt.plot([0, 1], [0, 1], 'k--')
        plt.xlabel('Tasa de Falsos Positivos (FPR)')
        plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
        plt.title(f"Curva ROC - {name} - Año {año}")
        plt.legend(loc="lower right")
        plt.show()

        # Exactitud y AUC
        accuracy = accuracy_score(y_test, y_pred)
        print(f"{name} - Año {año} - Exactitud: {accuracy:.4f}")
        print(f"{name} - Año {año} - AUC: {roc_auc:.4f}")

# Evaluar los modelos para cada año
entrenar_y_evaluar_modelo(X_train_2004, X_test_2004, y_train_2004, y_test_2004, 2004)
entrenar_y_evaluar_modelo(X_train_2024, X_test_2024, y_train_2024, y_test_2024, 2024)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
import pandas as pd
import matplotlib.pyplot as plt

# Guardar nombres de las columnas
columnas_2004 = df_2004.drop(columns=['ANO4', 'DESOCUPADOS']).columns
columnas_2024 = df_2024.drop(columns=['ANO4', 'DESOCUPADOS']).columns

# Separar las variables predictoras y objetivo
X_2004 = df_2004.drop(columns=['ANO4', 'DESOCUPADOS']).values
y_2004 = df_2004['DESOCUPADOS'].values

X_2024 = df_2024.drop(columns=['ANO4', 'DESOCUPADOS']).values
y_2024 = df_2024['DESOCUPADOS'].values

# Imputar los valores faltantes con la media
imputer = SimpleImputer(strategy='mean')
X_2004 = imputer.fit_transform(X_2004)
X_2024 = imputer.transform(X_2024)

# Definir y entrenar los modelos
model_2004 = LogisticRegression(penalty='l1', solver='liblinear', C=1.0)
model_2024 = LogisticRegression(penalty='l1', solver='liblinear', C=1.0)

model_2004.fit(X_2004, y_2004)
model_2024.fit(X_2024, y_2024)

# Función para mostrar las 6 variables más importantes con coeficientes != 0
def mostrar_importancia(model, columnas, year):
    coef = model.coef_[0]  # Coeficientes del modelo
    importancia = pd.DataFrame({'Variable': columnas, 'Coeficiente': coef})

    # Filtrar variables con coeficiente diferente de 0 y ordenar por magnitud
    importancia = importancia[importancia['Coeficiente'] != 0]
    importancia = importancia.sort_values(by='Coeficiente', key=abs, ascending=False)

    # Seleccionar solo las 6 más importantes
    importancia = importancia.head(6)

    # Gráfico de barras horizontal
    plt.figure(figsize=(10, 6))
    plt.barh(importancia['Variable'], importancia['Coeficiente'], color='teal')
    plt.xlabel('Coeficiente')
    plt.title(f'6 Variables Más Importantes - Año {year}')
    plt.gca().invert_yaxis()
    plt.show()

# Mostrar importancia para cada año
mostrar_importancia(model_2004, columnas_2004, 2004)
mostrar_importancia(model_2024, columnas_2024, 2024)

```
