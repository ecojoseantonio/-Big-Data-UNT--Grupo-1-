```python
# Nos interesa generar la variable proporción de menores de 16 años por adulto en el hogar
# Creamos columnas que indican si el individuo es menor a 16 años (menor) o mayor (adulto)
df['menor'] = df['CH06'] < 16
df['adulto'] = df['CH06'] >= 16

# Agrupamos por hogar y contamos menores y adultos
cuentas_por_hogar = df.groupby(['CODUSU', 'ANO4']).agg(
    num_menores=('menor', 'sum'),
    num_adultos=('adulto', 'sum')
).reset_index()

# Calculamos la proporción de menores por adultos
cuentas_por_hogar['proporcion_menores_adultos'] = (
    cuentas_por_hogar['num_menores'] / cuentas_por_hogar['num_adultos'].replace(0, float('nan'))
)

print(cuentas_por_hogar)
```

```python
menores_a_1 = (cuentas_por_hogar['proporcion_menores_adultos'] < 1).sum()
mayores_a_1 = (cuentas_por_hogar['proporcion_menores_adultos'] > 1).sum()

print(f"Valores menores a 1: {menores_a_1}")
print(f"Valores mayores a 1: {mayores_a_1}")
```

```python
# La siguiente variable que nos interesa generar es el nivel educativo máximo que alzancó un miembro del hogar
# Esta variable indica de algún modo el máximo nivel educativo del hogar
max_nivel_por_hogar = df.groupby(['CODUSU', 'ANO4'])['NIVEL_ED'].max().reset_index(name='max_nivel_ed')

print(max_nivel_por_hogar)
```

```python
# Contamos cuántos hogares hay para cada valor de max_nivel_ed, distinguiendo por año
conteo_niveles = (max_nivel_por_hogar
                  .groupby(['max_nivel_ed', 'ANO4'])
                  .size()
                  .reset_index(name='num_hogares'))

# Ver el resultado
print(conteo_niveles)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filtramos los datos para los años 2004 y 2024
data_2004 = conteo_niveles[conteo_niveles['ANO4'] == 2004].copy()
data_2024 = conteo_niveles[conteo_niveles['ANO4'] == 2024].copy()

# Definimos un diccionario para mapear los niveles educativos
nivel_educativo = {
    1: 'Primaria Incompleta',
    2: 'Primaria Completa',
    3: 'Secundaria Incompleta',
    4: 'Secundaria Completa',
    5: 'Superior Universitaria Incompleta',
    6: 'Superior Universitaria Completa',
    7: 'Sin instrucción'
}

# Creamos nuevas columnas con las descripciones
data_2004['nivel_descripcion'] = data_2004['max_nivel_ed'].map(nivel_educativo)
data_2024['nivel_descripcion'] = data_2024['max_nivel_ed'].map(nivel_educativo)

# Calculamos la frecuencia porcentual
data_2004.loc[:, 'frecuencia'] = (data_2004['num_hogares'] / data_2004['num_hogares'].sum()) * 100
data_2024.loc[:, 'frecuencia'] = (data_2024['num_hogares'] / data_2024['num_hogares'].sum()) * 100

plt.figure(figsize=(12, 6))

# Histograma para el año 2004
plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, gráfico 1
plt.barh(data_2004['nivel_descripcion'], data_2004['frecuencia'], color='skyblue', edgecolor='black')
plt.title('Frecuencia porcentual de hogares por nivel educativo (2004)')
plt.xlabel('Frecuencia porcentual')
plt.ylabel('Máximo nivel educativo')

# Histograma para el año 2024
plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, gráfico 2
plt.barh(data_2024['nivel_descripcion'], data_2024['frecuencia'], color='lightgreen', edgecolor='black')
plt.title('Frecuencia porcentual de hogares por nivel educativo (2024)')
plt.xlabel('Frecuencia porcentual')
plt.ylabel('Máximo nivel educativo')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Definir un diccionario para mapear los niveles educativos
nivel_educativo = {
    1: 'Primaria Incompleta',
    2: 'Primaria Completa',
    3: 'Secundaria Incompleta',
    4: 'Secundaria Completa',
    5: 'Superior Universitaria Incompleta',
    6: 'Superior Universitaria Completa',
    7: 'Sin instrucción'
}

# Crear un histograma en subgráficas
years = df['ANO4'].unique()
num_years = len(years)

plt.figure(figsize=(12, 6))

# Iterar sobre los años y crear un histograma para cada uno
for i, year in enumerate(years, start=1):
    plt.subplot(1, num_years, i)  # Crea una subgráfica
    subset = df[df['ANO4'] == year]
    
    # Calcular el total y el número de ocupados por nivel educativo
    total_counts = subset['NIVEL_ED'].value_counts()
    ocupados_counts = subset[subset['ESTADO'] == 1]['NIVEL_ED'].value_counts()

    # Asegurarse de que ambos contadores tengan el mismo índice
    total_counts = total_counts.reindex(range(1, 8), fill_value=0)
    ocupados_counts = ocupados_counts.reindex(range(1, 8), fill_value=0)

    # Graficar las barras horizontales
    bar_width = 0.35
    y = total_counts.index

    plt.barh(y - bar_width/2, total_counts, height=bar_width, label='Total', alpha=0.7, color='lightseagreen', edgecolor='black')
    plt.barh(y + bar_width/2, ocupados_counts, height=bar_width, label='Ocupados', alpha=0.7, color='lightcoral', edgecolor='black')

    # Ajustar el título y etiquetas
    plt.title(f'Nivel Educativo - {year}')
    plt.xlabel('Número de Personas')
    plt.ylabel('Nivel Educativo')
    plt.yticks(ticks=range(1, 8), labels=nivel_educativo.values())
    plt.grid(axis='x', linestyle='None', alpha=0.7)
    plt.legend()

plt.tight_layout()  
plt.show()
```

```python
import pandas as pd

# Definir los mapeos
estado_map = {
    0: 'Entrevista no realizada',
    1: 'Ocupado',
    2: 'Desocupado',
    3: 'Inactivo',
    4: 'Menor de 10 años'
}

nivel_educativo_map = {
    7: 'Sin instrucción',
    1: 'Primaria Incompleta',
    2: 'Primaria Completa',
    3: 'Secundaria Incompleta',
    4: 'Secundaria Completa',
    5: 'Superior Universitaria Incompleta',
    6: 'Superior Universitaria Completa',
    9: 'Ns./Nr.'
}

# Filtrar los datos para excluir ESTADO = 0 y NIVEL_ED = 9
filtered_df = df[(df['ESTADO'] != 0) & (df['ESTADO'] != 4) & (df['NIVEL_ED'] != 9)].copy()

# Convertir NIVEL_ED a un tipo categórico con un orden específico
filtered_df['NIVEL_ED'] = pd.Categorical(
    filtered_df['NIVEL_ED'].map(nivel_educativo_map), 
    categories=list(nivel_educativo_map.values()), 
    ordered=True
)

# Crear la tabla de contingencia
contingency_table = pd.crosstab(
    filtered_df['NIVEL_ED'],
    filtered_df['ESTADO'].map(estado_map),
    margins=True,
    margins_name='Total'
)

# Estilizar la tabla
styled_table = contingency_table.style \
    .background_gradient(cmap='Blues', axis=None) \
    .set_table_attributes('style="width: 50%; margin: auto; border-collapse: collapse;"') \
    .set_caption("Tabla de Contingencia: Estado vs Nivel Educativo") \
    .set_properties(**{'text-align': 'center'}) \
    .set_table_styles([{
        'selector': 'th',
        'props': [('background-color', '#f0f0f0'), ('font-weight', 'bold')]
    }, {
        'selector': 'td',
        'props': [('border', '1px solid #ddd'), ('padding', '8px')]
    }])

# Mostrar la tabla
styled_table
```
