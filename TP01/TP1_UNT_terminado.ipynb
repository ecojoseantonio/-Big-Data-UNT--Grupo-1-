Es decir, que las cadenas de caracteres finales tendrán la forma: https://www.lanacion.com.ar/*texto_obtenido*)

Nota 2: junto con su entrega, adjunten una captura de la página de noticias al momento de correr su código. Eso servirá al momento de la corrección para verificar que los links obtenidos hacen referencia a las noticias de ese día y hora.

```python
import requests #html requestor
from bs4 import BeautifulSoup #html parser
import pandas as pd #dataframe manipulator
```

```python
url = "https://www.lanacion.com.ar/"


r = requests.get(url) #traigo el contenido del html
contenido = r.content

soup = BeautifulSoup(contenido, "html.parser")
soup
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
news_elements = soup.find_all('article', class_='ln-card')

for news in news_elements:
    title_element = news.find('h2', class_='title')
    title = title_element.text.strip() if title_element else None

    link_element = news.find('a', class_='link')
    link = link_element['href'] if link_element else None

    # ... Extraer otros datos de la noticia ...

    print(f"Título: {title}")
    print(f"Enlace: {link}")
```

```python
news_data = []
for news in news_elements:
    title_element = news.find('h2', class_='title')
    title = title_element.text.strip() if title_element else None

    link_element = news.find('a', class_='link')
    link = link_element['href'] if link_element else None

    news_data.append({'Título': title, 'Enlace': link})

df = pd.DataFrame(news_data)
print(df)
```

```python
df.to_excel('noticias.xlsx', index=False)
```
